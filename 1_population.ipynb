```python
# 일별 유동인구수 조사
a = [242, 256, 237, 263, 81, 46]
print('A=',a)
```

```python
n = len(a)
my_sum = 0
my_avg = 0

for i in range(0,n):
    my_sum = my_sum + a[i]
    
my_avg = my_sum/n
print('Total sum: ', my_sum)
print('Total avg: ', my_avg)
```

```python
# 시간대를 세분화해서 분석
import pandas as pd

data = pd.read_csv('./passby_data.CSV')
data.head()
```

```python
data1 = [] # 빈 리스트

end = int(len(data)/24)

for i in range(0, end):    
    n1 = i*24
    n2 = (i+1)*24
    data1.append(data[n1:n2])

```

```python
data.info()
```

```python
import numpy as np
time = np.linspace(0,23,24)
len(time)
```

```python
day = ['mon','tue','wed','thr','fri','sat','sun']

n = 0
plt.figure(figsize=(12,12))
for i in range(0,7):
    for j in range(0,3):
        n = n + 1
        plt.subplot(7,3,n)
        plt.plot(time, data1[i][data.columns[j]])
        plt.ylabel(day[i])
        plt.xlabel('time(hr)')
        plt.title(data.columns[j])
     
plt.show()
```

```python
lname = ['pass','women','below 30']

n = 0
plt.figure(figsize=(12,12))
for i in range(0,7):
    n = n + 1
    plt.subplot(7,1,n)
    plt.plot(time, data1[i][data.columns])
        
    plt.ylabel(day[i])
    plt.xlabel('time(hr)')   
    plt.xticks(np.arange(0,24,6))
    plt.legend(lname)
    plt.grid()
     
plt.show()
```
